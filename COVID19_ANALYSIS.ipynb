```python
covid_data['Date']=pd.to_datetime(covid_data['Date'],format='%Y-%m-%d')
covid_data.head(15)
```

```python
covid_data['Active_cases']=covid_data['Confirmed']-(covid_data['Cured']+covid_data['Deaths'])
covid_data.tail()
```

```python
covid_data['State/UnionTerritory']=covid_data['State/UnionTerritory'].replace('Maharashtra***',"Maharashtra")
covid_data['State/UnionTerritory']=covid_data['State/UnionTerritory'].replace('Bihar****',"Bihar")
covid_data['State/UnionTerritory']=covid_data['State/UnionTerritory'].replace('Madhya Pradesh***',"Madhya Pradesh")
covid_data['State/UnionTerritory']=covid_data['State/UnionTerritory'].replace('Karanataka',"Karnataka")
```

```python
statewise_data=pd.pivot_table(covid_data,values=
                              ['Confirmed','Deaths','Cured'],
                              index='State/UnionTerritory', aggfunc=max)
```

```python
statewise_data['Recovery Rate']=statewise_data['Cured']*100/statewise_data['Confirmed']
```

```python
statewise_data['Mortality Rate']=statewise_data['Deaths']*100/statewise_data['Confirmed']
statewise_data=statewise_data.sort_values(by="Confirmed",ascending=False)
```

```python
statewise_data.style.background_gradient(cmap="CMRmap")
```

```python
top10ActiveCases=covid_data.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(15,10))
plt.title("Top 10 states with most Active Cases  in India",size=22)
ax=sns.barplot(data=top10ActiveCases.iloc[:10],y='Active_cases',x='State/UnionTerritory',linewidth=2,edgecolor='black')
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
top10Deaths=covid_data.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
```

```python
vaccine=vaccination_s[vaccination_s.State!='India']
vaccination_s.head()
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated':"Total"},inplace=True)
vaccination_s.head()
```

```python
max_vacc=vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vacc=max_vacc.sort_values(by='Total',ascending=False)[:5]
```

```python
fig=plt.figure(figsize=(12,5))
plt.title("Top 5 Vaccinated States in India",size=16)
x=sns.barplot(data=max_vacc.iloc[:10],y=max_vacc.Total,x=max_vacc.index,linewidth=2,edgecolor='black')
```

```python
min_vacc=vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vacc=min_vacc.sort_values(by='Total',ascending=True)[:5]
```

```python
fig=plt.figure(figsize=(15,7))
plt.title("Least 5 Vaccinated States in India",size=16)
x=sns.barplot(data=max_vacc.iloc[:10],y=min_vacc.Total,x=min_vacc.index,linewidth=2,edgecolor='red')
```
